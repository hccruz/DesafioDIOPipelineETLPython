# Importar a Biblitoteca

```python
import pandas as pd
```

# Extração

Extração dos dados em .csv

```python
contratos = pd.read_csv('tabela_contratos.csv')
```

```python
contratos
```

```python
datas = pd.read_csv('tabela_datas.csv')
```

```python
datas
```

```python
empresas = pd.read_csv('tabela_empresas.csv')
```

```python
empresas
```

#Transformação



```python
contratos_empresa = contratos.merge(empresas,
                                    left_on='fk_empresa_contratada',
                                    right_on='id_empresa',
                                    how='left')
```

```python
contratos_empresa
```

```python
contratos_empresa.drop(columns=['fk_empresa_contratada', 'id_empresa'], inplace=True)
```

```python
contratos_empresa.head()
```

```python
contratos_final = contratos_empresa.merge(datas,
                                    left_on='inicio_vigencia',
                                    right_on='id_data',
                                    how='left')
```

```python
contratos_final
```

```python
contratos_final.drop(columns=['inicio_vigencia', 'id_data'], inplace=True)
```

```python
contratos_final.head()
```

```python
contratos_final.rename(columns={'data': 'data_ini_vigencia'}, inplace=True)
```

```python
contratos_final.head()
```

```python
contratos_final = contratos_final.merge(datas,
                                    left_on='termino_vigencia',
                                    right_on='id_data',
                                    how='left')
```

```python
contratos_final.drop(columns=['termino_vigencia', 'id_data'], inplace=True)
```

```python
contratos_final.head()
```

```python
contratos_final.rename(columns={'data': 'data_ter_vigencia'}, inplace=True)
```

```python
contratos_final.head()
```

```python
contratos_final.dtypes
```

```python
contratos_final.data_ter_vigencia = contratos_final.data_ter_vigencia.str.replace('31/09/2017', '30/09/2017')
```

```python
contratos_final.data_ini_vigencia = pd.to_datetime(contratos_final.data_ini_vigencia, format='%d/%m/%Y').dt.date
contratos_final.data_ter_vigencia = pd.to_datetime(contratos_final.data_ter_vigencia, format='%d/%m/%Y').dt.date
```

```python
contratos_final.head()
```

```python
contratos_final['tempo_contrato'] = (contratos_final.data_ter_vigencia - contratos_final.data_ini_vigencia).dt.days
```

```python
contratos_final.head()
```

# Load

Carregar os dados como uma planilha do Excel

```python
contratos_final.to_excel('contratosfinal.xlsx')
```
